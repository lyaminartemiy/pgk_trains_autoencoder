```python
import torch
import cv2
import numpy as np
from matplotlib import pyplot as plt
from torch.utils.data import Dataset
from torchvision import transforms
from PIL import Image
from IPython.display import display

from src.dataset.create_dataset import TrainsDataset
from src.dataset.main_dataset import main_dataset
from src.config.config import config


train_dataset = main_dataset()
```

```python
def get_image(id_image):
    img = train_dataset[id_image].permute(1, 2, 0)
    img = img.detach().numpy()
    img = img.astype(np.uint8)
    img = Image.fromarray(img)
    return img


def scale_image(img, scale=10):
    img = img.resize((int(img.width * scale), int(img.height * scale)))
    return img
```

```python
image_n = 5
scale = 4

for image_id in range(image_n):
    img = get_image(image_id + 15)
    img = scale_image(img, scale)
    display(img)
```
